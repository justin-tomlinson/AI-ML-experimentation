# Colour Conversion and Geometric Transformations

This workbook i experiemnt with various colour conversions using OpenCV and then go on to define and apply several transforamtions. Frist up though i need to import the required libraries.

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

Then I can read in the sample image

```python
img = cv.imread('./resources/img1.jpg')
```

## 1.Colour Conversion

### Convert the image colour space to HSV Space
cvtcolor() call can convert colour ranges for an image

```python
img_hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
cv.imshow('image in HSV', img_hsv)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.figure(figsize=(12,9))
plt.imshow(cv.cvtColor(img_hsv, cv.COLOR_BGR2RGB))
plt.axis("off")
```

```python
cv.imwrite('image1_hsv.png', img_hsv)
```

### Convert the image colour space to Grayscale
cvtcolor() call can convert colour ranges for an image including grayscale

```python
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv.imshow('image in gray', img_gray)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.figure(figsize=(12,9))
plt.imshow(img_gray,'gray')
plt.axis("off")
```

```python
cv.imwrite('image1_gray.png', img_gray)
```

## 2. Geometric Transformations

### 2.1 Scaling
Images can be scaled

```python
height, width = img.shape[:2]
h_scale = 0.5
v_scale = 0.4
new_height = (int) (height * v_scale) # we need this as the new height must be interger
new_width = (int) (width * h_scale) # we need this as the new width must be interger
```

```python
img_resize = cv.resize(img, (new_width, new_height), interpolation = cv.INTER_LINEAR)
cv.imshow('resize', img_resize)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.figure(figsize=(12,9))
plt.imshow(cv.cvtColor(img_resize, cv.COLOR_RGB2BGR))
plt.axis("off")
```

```python
cv.imwrite('image1_resize.png', img_resize)
```

### 2.2 Translation

```python
t_x = 100
t_y = 200

M = np.float32([[1, 0, t_x], [0, 1, t_y]])

height, width = img.shape[:2] #this will get the number of rows and columns in img

img_translation = cv.warpAffine(img, M, (width, height))
cv.imshow('translation', img_translation)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.figure(figsize=(12,9))
plt.imshow(cv.cvtColor(img_translation, cv.COLOR_RGB2BGR))
plt.axis("off")
```

```python
cv.imwrite('image1_translation.png', img_translation)
```

### 2.3 Rotations
Imags can be rotated

```python
theta = 45 #rotate 45 degrees in anti-clockwise

c_x = (width - 1) / 2.0 # column index varies in [0, width-1]
c_y = (height - 1) / 2.0 # row index varies in [0, height-1]
c = (c_x, c_y) # A point is defined by x and y coordinate
print(c)
```

```python
s = 1
M = cv.getRotationMatrix2D(c, theta, s)

img_rotation = cv.warpAffine(img, M, (width, height))
cv.imshow('rotation', img_rotation)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.figure(figsize=(12,9))
plt.imshow(cv.cvtColor(img_rotation, cv.COLOR_RGB2BGR))
plt.axis("off")
```

```python
cv.imwrite('image1_rotation.png', img_rotation)
```

### 2.4 Affine Trransformations

```python
m00 = 0.38
m01 = 0.27
m02 = -47.18
m10 = -0.14
m11 = 0.75
m12 = 564.32
M = np.float32([[m00, m01, m02], [m10, m11, m12]])
```

```python
height, width = img.shape[:2]
img_affine = cv.warpAffine(img, M, (width, height))
cv.imshow('affine', img_affine)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
plt.figure(figsize=(12,9))
plt.imshow(cv.cvtColor(img_affine, cv.COLOR_RGB2BGR))
plt.axis("off")
```

```python
cv.imwrite('image1_affine.png', img_affine)
```

# Basic IO with OpenCV
This workbook was a created for one of my university classes. It was really just to document and experiment with the OpenCV basic functions. Its super simple but used as a reference.

Import the required libraries. In this case we are using OpenCV to read the images and MatplotLib for displaying them within the workbook.

```python
import cv2 as cv
from matplotlib import pyplot as plt
```

Read the sample image.

```python
img = cv.imread('./resources/img1.jpg')
```

Get the Dimensions of the Image can be retreived as a tuple from the img.shape call

```python
height, width = img.shape[:2]
print(f'The height is {height} and width is {width}')
```

Displaying the image in a pop out window using openCV imshow. 

**Press any key to close the window.**

```python
cv.imshow('input image', img)

# These are important so that the window closes when you press any key
cv.waitKey(0)
cv.destroyAllWindows()
```

The image can also be written to disk using the command below

```python
cv.imwrite('image1_out.png', img)
```

Sometimes you might want to display the image inside the notebook. The Matplotlib librarty can be used to plot this. Use the figure(figsize=()) call to change the display size.

Be sure to define the colour channel as matplotlibs defaults are not the same as OpenCV.

```python
plt.figure(figsize=(12,9))
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.axis("off")
```
